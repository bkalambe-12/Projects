```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine=load_wine()
X= wine.data[:,2].reshape(-1,1)
Y= wine.target.reshape(-1,1)

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=20)

```

```python
mean_X = np.mean(X_train)
mean_Y = np.mean(Y_train)

numerator = np.sum((X_train - mean_X)*(Y_train - mean_Y))
denominator = np.sum((X_train - mean_X)**2)
beta = numerator/ denominator
alpha = mean_Y - beta*mean_X

Y_train_pred= alpha+beta*X_train
Y_test_pred= alpha+beta*X_test
print('Himanshu Kothari: 21CE1376')
```

```python
plt.scatter(X_train, Y_train)
plt.plot(X_train,Y_train_pred, color='red')
plt.xlabel('Chemical Properties')
plt.ylabel('Alcohol Content')
plt.title('Linear Regression on Digits Dataset(Training Set)')
plt.show()
```

```python
plt.scatter(X_test, Y_test)
plt.plot(X_test, Y_test_pred, color='green')
plt.xlabel('Chemical Properties')
plt.ylabel('Alcohol Content')
plt.title('Linear Regression on Wine Dataset(Testing Set)')
plt.show()
```
